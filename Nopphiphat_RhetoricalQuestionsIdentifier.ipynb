1) And I said, "Why won't everybody go south?" They say, "It would be disruptive."

2) I said, "For how long?" I finally got them up for 12 to 15 years.

```python
import re, nltk
from os import listdir
from os.path import isfile, join
from nltk import data
from nltk.util import bigrams
from nltk.tokenize import TreebankWordTokenizer, word_tokenize
nltk.download('punkt')
sentence_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
treebank_tokenizer = TreebankWordTokenizer()

dir_base = "/Users/HyunjaeCho/Documents/GitHub/f19_ds_nlp/homeworks/homework_1/data/"
```

```python
def read_file(filename):
    input_file_text = open(filename, encoding = 'utf-8').read()
    return input_file_text

def read_directory_files(directory):
    file_texts = []
    files = [f for f in listdir(directory) if isfile(join(directory, f))]
    #print(files)
    for f in files:
        file_text = read_file(join(directory, f))
        #print(file_text)
        file_texts.append({"file": f, "content": file_text})
    return file_texts
```

Read all files into dictionary within list.
Each element in the list is a dictionary with two elements: file, content

which can be accessed by: list[dictionary_number][dictionary_element]. 

1) text_corpus[0] - first dictionary element of list 'text_corpus'

2) text_corpus[0][file] - first dictionary element's file of list 'text_corpus'

3) text_corpus[0][content] - first dictionary element's content of list 'text_corpus'

```python
text_corpus = read_directory_files(dir_base)
print(text_corpus)
```

```python
def process_debate(debate_object):
    
    number_of_debates = len(debate_object)
    list_of_debates = []
    possible_rhetorical_questions = []
    all_words = []
    for i in range(number_of_debates):
        debate_name = debate_object[i]
        debate_text = debate_name["content"]
        list_of_debates.append(debate_object[i]["content"])
    
    all_sentences = list_of_debates[0] + " " + list_of_debates[1] + " " +  list_of_debates[2] + " " + list_of_debates[3] + " " + list_of_debates[4] + " " + list_of_debates[5]
    tokenized_words = treebank_tokenizer.tokenize(all_sentences)
    tokenized_sentences = sentence_tokenizer.tokenize(all_sentences)
    
    for i in tokenized_sentences:
        check = re.findall(r'\s*([^.?]*(?:[hH]ow|[cC]an|[wW]hat|[wW]here|[dD]escribe|[wW]ho|[wW]hen)[^.?]*?\s*\?)', i)
        for s in check:
            possible_rhetorical_questions.append(s)
    
    return possible_rhetorical_questions, tokenized_sentences, tokenized_words, all_sentences
```

```python
# Process the debates and output possible_rhetorical_questions, tokenized_sentence, tokenized_words, and all_sentences
possible_rhetorical_questions, tokenized_sentences, tokenized_words, all_sentences = process_debate(text_corpus)
```

```python
# Remove duplicates within list of possible rhetorical questions
temp = []
for i in possible_rhetorical_questions:
    if i not in temp:
        temp.append(i) 
num_possible_rhetorical_questions = temp
print(len(num_possible_rhetorical_questions))
```

```python
# Find the indexes for all the sentences within tokenized_sentences
# that match all questions inside the possible_rhetorical_questions
possible_rhetorical_questions_index = []
for question in possible_rhetorical_questions:
    if question in tokenized_sentences:
        x = tokenized_sentences.index(question)
        possible_rhetorical_questions_index.append(x)
print(len(possible_rhetorical_questions_index))
print(possible_rhetorical_questions_index)
#print(debate_sentences[26])
```

```python
candidate_rhetorical_questions_index = []
for i in possible_rhetorical_questions_index:
    prev_sentence = tokenized_sentences[i-1] 
    next_sentence = tokenized_sentences[i+1]
    #print(prev_sentence + "\n")
    if ((re.match('^[A-Z][^?!.]*[.!]$', next_sentence)) and (re.match('^(?![A-Z]+(\s)[A-Z]+\.)', next_sentence)) and (re.match('^(?!([A-Z][a-z]+|[A-Z]+)\.)', next_sentence)) and (re.match('^(?![A-Z][a-z]+(\s)+[A-Z][a-z]+\.)', next_sentence)) and (re.match('^(?![A-Z]+\:)', next_sentence)) and (re.match('^[A-Z][^?!.]*[.!]$', prev_sentence))):        
        candidate_rhetorical_questions_index.append(i)
        #print(i)
print(len(candidate_rhetorical_questions_index))
print(candidate_rhetorical_questions_index)
```

```python
clean_rhetorical_questions = []
for i in candidate_rhetorical_questions_index:
    #print(tokenized_sentences[i] + "\n")
    #print(tokenized_sentences[i-1] + " " + tokenized_sentences[i] + " " + tokenized_sentences[i+1] + "\n")
    clean_rhetorical_questions.append(tokenized_sentences[i])
    
print(clean_rhetorical_questions)
```

```python
speaker_sentences = []
for i in range(num_tokenized_sentences):
    for name in speakers:
        if name in tokenized_sentences[i]:
            #print(tokenized_sentences[i])
            speaker_sentences.append(i)
print("Indexes of sentences with speakers: " + "\n")
print(speaker_sentences)
print("\n" + "Indexes of sentences that are rhetorical questions" + "\n")
print(candidate_rhetorical_questions_index)
```

```python
speaker_rhetorical = []
for i in range(len(candidate_rhetorical_questions_index)):
    speaker_sentence_index = min(speaker_sentences, key=lambda x:abs(x-candidate_rhetorical_questions_index[i]))
    speaker_sentence = tokenized_sentences[speaker_sentence_index]
    #print(speaker_sentence)
    #print(tokenized_sentences[candidate_rhetorical_questions_index[i]])
    for name in speakers:
        if name in speaker_sentence:
            speaker_rhetorical.append({"speaker": name, "questions": tokenized_sentences[candidate_rhetorical_questions_index[i]]})
#print(speaker_rhetorical)
print(str(len(speaker_rhetorical)) + " rhetorical questions" + "\n")

for i in range(len(speaker_rhetorical)):
    print(speaker_rhetorical[i])
    print("\n")
```
